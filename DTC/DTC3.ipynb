```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from scipy.stats import zscore
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Bank_Personal_Loan_Modelling_train.csv')
data_test = pd.read_csv('Bank_Personal_Loan_Modelling_reserved.csv')
```

```python
data.drop('ID', axis=1, inplace=True)
data.corr().abs().sum().idxmin()
data.drop(data.corr().abs().sum().idxmin(), axis=1, inplace=True)
```

```python
data['Experience'] -= data['Experience'].min()
data['CCAvg'] *= 12
```

```python
data_test.drop('ID', axis=1, inplace=True)
data_test.corr().abs().sum().idxmin()
data_test.drop(data_test.corr().abs().sum().idxmin(), axis=1, inplace=True)
data_test['Experience'] -= data_test['Experience'].min()
data_test['CCAvg'] *= 12
```

```python
x_train = data.drop('Personal Loan', axis=1)
y_train = data['Personal Loan']
x_test = data_test
```

```python
tree = DecisionTreeClassifier(random_state=17, criterion='gini', class_weight='balanced')
tree.fit(x_train, y_train)
y_pred = tree.predict(x_test)
print(list(y_pred))
```
