```python
from keras.datasets import mnist
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

(_,_), (x, y) = mnist.load_data()

x = np.array([i.flatten() for i in x])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=15)
km = KMeans(random_state=15, n_clusters=10)
km.fit(x_train, y_train)
len([i for i in km.labels_ if i == 6])
```

```python
data = pd.DataFrame({'cluster_label': km.labels_, 'true_label': y_train})
data = data.groupby('cluster_label')['true_label'].value_counts().rename('count').reset_index()

labels = []

for i in range(10):
    labels.append(data[data['cluster_label'] == i].head(1)['true_label'].values[0])

', '.join([str(i) for i in labels])
```

```python
accuracy_score(y_train, [labels[i] for i in km.labels_])
```

```python
y_pred = km.predict(x_test)
accuracy_score(y_test, [labels[i] for i in y_pred])
```

```python
confusion_matrix(y_test, [labels[i] for i in y_pred])[9].argmax()
```

```python
tsne = TSNE(n_components=2, init='random', random_state=15)
x_train_ = tsne.fit_transform(x_train)
```
