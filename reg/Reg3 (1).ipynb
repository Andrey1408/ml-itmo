```python
import pandas as pd
import seaborn as sns
import numpy as np
import io
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import Ridge
```

```python
df_train = pd.read_csv("fish_train.csv")
df_test = pd.read_csv("fish_reserved.csv")
```

```python
cat_mask_train = (df_train.dtypes.values==object)
cat_mask_test = (df_test.dtypes.values==object)
df_train1 = pd.get_dummies(df_train[df_train.columns[cat_mask_train]], drop_first=True)
df_train_new = pd.concat([df_train, df_train1],axis=1)
df_test1 = pd.get_dummies(df_test[df_test.columns[cat_mask_test]], drop_first=True)
df_test_new = pd.concat([df_test, df_test1], axis=1)
df_train_new = df_train_new.drop(columns='Species')
df_test_new = df_test_new.drop(columns='Species')
df_train_ = df_train_new.dropna()
df_test_ = df_test_new.dropna()
```

```python
x = df_train_.drop(columns='Weight',axis=1)
y = df_train_['Weight']
x
```

```python
pca = PCA(n_components=1, svd_solver='full')
pca.fit(df_train[['Length1', 'Length2', 'Length3']])
```

```python
x['Lengths'] = pca.fit_transform(x[['Length1', 'Length2', 'Length3']])[:, 0]
```

```python
x = x.drop(columns='Length1')
x = x.drop(columns='Length2')
x = x.drop(columns='Length3')
```

```python
df_test_['Lengths'] = pca.fit_transform(df_test_[['Length1', 'Length2', 'Length3']])[:, 0]
```

```python
df_test_ = df_test_.drop(columns='Length1')
df_test_ = df_test_.drop(columns='Length2')
df_test_ = df_test_.drop(columns='Length3')
```

```python
x[['Height','Width']] = x[['Height','Width']].apply(lambda x: x**1.7)
df_test_[['Height','Width']] = df_test_[['Height','Width']].apply(lambda x: x**1.7)
x['Lengths'] = x['Lengths']**2
df_test_['Lengths'] = df_test_['Lengths']**2
```

```python
reg = LinearRegression().fit(x, y)
Y_pred = reg.predict(df_test_)
print(list(Y_pred))
```
